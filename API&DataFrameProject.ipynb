```python
import requests
import json

# Please enter your own Open Weather API key
key = 'XXXXXXXXXXXXXX'

units = 'metric'

# List needed cities
cities = ['New+York', 'New+Jersey','Chicago', 'London', 'Dubai', 'Shanghai', 'Sydney','Miami','San', 'Seattle']

test = 'New+York'

# Prepare empty lists for needed info: Max/Min Tempeartur, Cloudiness, Humidity and Wind Speed
max_temp=[]
min_temp=[]
Cloudiness=[]
humidity=[]
wind_speed=[]
```

```python
# Create a function to scrap requests info
def datainfo(n):
  url = f'http://api.openweathermap.org/data/2.5/weather?q={n}&units={units}&appid={key}'
  r = requests.get(url)
  data = r.json()
  json.dumps(data, sort_keys=True, indent = 6)
  max_temp.append(data['main']['temp_max'])
  min_temp.append(data['main']['temp_min'])
  Cloudiness.append(data['clouds']['all'])
  humidity.append(data['main']['humidity'])
  wind_speed.append(data['wind']['speed'])
  # Save as a dictionary
  data_dict = {
      'Max Temp': max_temp,
      'Min Temp': min_temp,
      'Cloudiness': Cloudiness,
      'Humidity': humidity,
      'Wind Speed': wind_speed
  }
  return data_dict
```

```python
def final(list):
  final_d = {}
  for i in list:
    final_d.update(datainfo(i))
  return final_d

final_dict = final(cities)

# Re-write Cities list
def new(list):
  City = []
  for x in list:
    if '+' in x:
      x = x.replace('+', ' ')
      City.append(x)
    else:
      City.append(x)
  return City

final_dict['City Name']=new(cities)

# Print the organized dictionary
final_dict
```

```python
# Create data frame 
import pandas as pd

df = pd.DataFrame(final_dict)

# Print the initial data frame
df
```

```python
# Move city name to the first column
cols = list(df)
cols.insert(0, cols.pop(cols.index('City Name')))
df = df.loc[:,cols]
df

# Storing data with the max temp
df.sort_values(by='Max Temp', ascending=False, inplace=True)

# Print the organized data frame
df.reset_index(drop=True, inplace=True)
df
```

```python
# Print the description statistics for the data frame
df.describe()
```

```python
# Save the data frame as csv.file
df.to_csv("10 Cities Weather Summary.csv")
```
