```python
import random
a = 0
b = 0
Name = []

def main(): 
    n = int(input("How many players? "))
    for i in range(1, n+1):
        Name.append(input("Player {}: ".format(i)))

    print("\n")
    
    file1 = open("Bowling.txt", 'a')
    for y in range (1, 11):
        print("Round {}".format(y))
        for x in range(n):
            a = random.randint(0, 10)
            if a != 10:
                b = random.randint(0, (10-a))   
                file1.write(Name[x]+'\n')  
                file1.write(str(a)+'\n')
                file1.write(str(b)+'\n')
            else:
                b = "/"
                file1.write(Name[x]+'\n')  
                file1.write(str(a)+'\n')
                file1.write(str(b)+'\n')

            print("{} : {}   {}".format(Name[x],a,b))
        print("\n")
    file1.close()
    return Name
```

```python
def Total(x):
    total = 0
    subtotal = 0
    
    file2 = open("Bowling.txt", 'r')
    person = file2.readline().rstrip()
    
    while person != '':
        a = file2.readline().rstrip()    
        b = file2.readline().rstrip()
        if person == x:
            if b != '/':
                subtotal = int(a)+int(b)
                total += subtotal
            else:
                subtotal = int(a)
                total += subtotal
        person = file2.readline().rstrip()
            
    file2.close()
    return total

for x in main():
    print("Total score of {} is: {} \n". format(x,Total(x)))
```
